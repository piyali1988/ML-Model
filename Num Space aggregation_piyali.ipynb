```python
import os
import numpy as np
import pandas as pd
import xlrd
```

```python
path = 'K:/New Reorganization/Research/TMW/TMW Longitudinal/Coding/Number_Space Coding/VS5_NS_Coding/Done/'
files = os.listdir(path)
#files
#os.remove('K:/New Reorganization/Research/TMW/TMW Longitudinal/Coding/Number_Space Coding/VS5_NS_Coding/Done/~$1775_VS5.xlsx')

# Iterate through all files and check for xlsx file extension
files_xls = [f for f in files if f[-4:]=='xlsx']
files_xls
```

```python
# read each excel file and put together in a dataframe
df = pd.DataFrame()
for f in files_xls:
    #print(path + f)
    data = pd.read_excel(path + f)
    df = df.append(data)
```

```python
# testing with one excel file first
# data = pd.read_excel('K:\\New Reorganization\\Research\\TMW\\TMW Longitudinal\\Coding\\Number_Space Coding\\VS5_NS_Coding\\Done\\262_VS5.xlsx')
```

```python
# Selecting target columns and replace any missing values with 0
data_subset = df[["File_VS", "Dimension", "Shape", "Feature_Property", "Location_Direction", 
                  "Amount","Pattern_Comparison", "cardinality"]].fillna(0)
data_subset.head()
```

```python
# Group the files based on idividual ids processed then Sum together.
grouped_df = data_subset.groupby(["File_VS"])
grouped_and_summed = grouped_df.sum()
grouped_and_summed = grouped_and_summed.reset_index()
# format for integer values only
pd.options.display.float_format = '{:.0f}'.format
grouped_and_summed.head()
```

```python
# Aggregated columns renamed
grouped_and_summed = grouped_and_summed.rename(columns={"Amount":"amount_sum","cardinality":"cardinality_sum", "Dimension":"dimension_sum", "Feature_Property":"feature_sum", "Location_Direction":"location_sum", "Pattern_Comparison":"pattern_comparison_sum", "Shape":"shape_sum"})
```

```python
# removing garbage values
aggregated_data = grouped_and_summed.iloc[2:]
# Arrange the columns alphabetically
aggregated_data = aggregated_data.sort_index(axis=1)
```
