```python
def drop_filler(x):
    x = str(x)
    x = x.strip()
    x = x.split()
    word_list = []
    stop_phrases = ['[Unintelligible',  'speech]', '[Coughs]', '[Crying]', '[Laughing]', '[Laughs]', '[Whining]', 
                    '[Gasps]', '[Whispering]', '[Shrieks]']
    for word in x:
        if word in stop_phrases:
            word.replace(word, '')
        else:
            word_list.append(word)
    return ' '.join(word_list)
```

```python
nli = []

for df in li:
    count = 0
    try:
        df[1]=df[1].map(lambda x: drop_filler(x))
        nli.append(df)
    except KeyError:
        continue
```
