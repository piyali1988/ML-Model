```python
import re 
import datetime
from datetime import datetime

def getage(field_date, dob):
    years = field_date.year - dob.year
    months = field_date.month - dob.month
    if field_date.day < dob.day:
        months -= 1
    if months < 0:
        months += 12
        years -= 1
    return '%sy%smo'% (years, months)
```

```python
def getage(now, dob):
    years = now.year - dob.year
    months = now.month - dob.month
    if now.day < dob.day:
        months -= 1
        while months < 0:
            months += 12
            years -= 1
    total_months = years * 12 + months
    return total_months
```
